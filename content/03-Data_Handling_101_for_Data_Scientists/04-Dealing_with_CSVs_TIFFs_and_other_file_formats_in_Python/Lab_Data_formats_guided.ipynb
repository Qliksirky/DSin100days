For the training images: <br>
1) The first four bytes of the file correspond to the magic number. This is an identifer for the file <br>
2) The next four bytes correspond to the number of images in the training set <br>
3) The next four bytes correspond to the x-dimension of the image <br>
4) The next four bytes correspond to the y-dimension of the image <br>
5) The rest of the bits in the file correspond the image data.  <br>
